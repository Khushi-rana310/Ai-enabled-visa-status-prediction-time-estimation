```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/visa_data.csv",
    encoding="latin1",
    sep=None,
    engine="python"
)

df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
```

```python
X = df.drop(columns=['visa_category_(label)'])
y = df['visa_category_(label)']
```

```python
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns

categorical_cols, numerical_cols
```

```python
X[categorical_cols].nunique().sort_values()
```

```python
low_cardinality_cols = [
    'marital_status',
    'education_level',
    'sponsorship_status',
    'financial_status',
    'language_proficiency',
    'purpose_of_travel',
    'visa_fee_payment_status',
    'medical_exam_status',
    'criminal_record_check'
]

high_cardinality_cols = [
    'nationality',
    'occupation',
    'target_country',
    'current_residence'
]
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=low_cardinality_cols,
    drop_first=True
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in high_cardinality_cols:
    X_encoded[col] = le.fit_transform(X_encoded[col])
```

```python
X_encoded.shape
X_encoded.head()
```

```python
if y.dtype == 'object':
    y = LabelEncoder().fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)
```
